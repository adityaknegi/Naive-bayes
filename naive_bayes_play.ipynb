```python
import pandas as pd 
import numpy as np
from pprint import pprint
```

```python
D={'D1':['Sunny','Hot','High','Weak','No'],'D2':['Sunny','Hot','High','Strong','No'],'D3':['Overcast','Hot','High','Weak','Yes'],
     'D4':['Rain','Mild','High','Weak','Yes'],
  'D5':['Rain','cool','Normal','Weak','Yes'],'D6':['Rain','cool','Normal','Strong','No'],
   'D7':['Overcast','cool','Normal','Strong','Yes'],'D8':['Sunny','Mild','High','Weak','No'],'D9':['Sunny','cool','Normal','Weak','Yes'],'D10':['Rain','Mild','Normal','Weak','Yes'],
   'D11':['Sunny','Mild','Normal','Strong','Yes'],
   'D12':['Overcast','Mild','High','Strong','Yes'],'D13':['Overcast','Hot','Normal','Weak','Yes'],
  'D14':['Rain','Mild','High','Strong','No']}
```

```python
df = pd.DataFrame(D,columns=['D1','D2','D3','D4','D5','D6','D7','D8','D9','D10','D11','D12','D13','D14'])
df=df.T
df.rename(columns={0:'Outlook',
                          1:'Temperature',2:'Humidity',
                    3:'Wind',4:'PlyTennis'}, 
                 inplace=True)
df
```

```python
class naive_bayes:
    
    def __init__(self):
        self.dic ={}
        self.class_unique = None
        
    def fit(self,df):
        self.class_unique = df[df.columns[-1]].unique()
        for col_name in df.columns:
            table=self.__likehood_table(df,col_name) 
            self.dic[col_name]=table
            #prob for class 
        for class_i in self.class_unique:
            self.dic[class_i] = sum(df[df.columns[-1]]==class_i)/(len(df)+0.0)

    
    def __likehood_table(self,df,col_name):
        col_name=col_name
        #unqiue class
        class_unique = df[df.columns[-1]].unique()
        #unique value in col
        unique_col_val=df[col_name].unique()

        dic_likehood_table={}
        #prior estimate(usually, p = 1/t for t possible values of xi
        p=1/(len(unique_col_val)+0.0)
        for class_i in class_unique:
            # play Yes or no tuples only
            small_df = df[df[df.columns[-1]]==class_i]
            dict_val={}
           
            for unique_val_i in unique_col_val:
                m=1 
                dict_val[unique_val_i]=(sum(small_df[col_name]==unique_val_i)+m*p)/(len(small_df)+0.0+m)
            dic_likehood_table[class_i]=dict_val
        # col is class , row is diff value in col
        # likehoood Outlook  sunny given play is Yes
        return dic_likehood_table
    
    def __testing(self,X_text,index_i):
        
        index_i=index_i
        _input = X_text.iloc[index_i]
        class_unique = self.class_unique
        p=[]
        #Decision making map rule
        for class_i in class_unique:
            total_prob=1
            # for Yes or No
            for (input_col_i,col_i) in zip(_input,X_text.columns):
                total_prob = total_prob*self.dic[col_i][class_i][input_col_i]
            #total probi 
            total_prob=self.dic[class_i]*total_prob
            p.append(total_prob)
    
        # max value index 
        index_class = class_unique[np.argmax(p)]
        return   index_class

    def prediction(self,X_text):    
        predict =[]
        for index_i in range(len(X_text)):
            predict.append(self.__testing(X_text,index_i))
        return predict
    
    def evaluate(self,X_test,y_test):
        value=self.prediction(X_test)
        print('accuracy :: {}'.format(sum(value==y_test)/(len(y_test)+0.0)))

Model=naive_bayes()
Model.fit(df)
print(Model.prediction(df[df.columns[:-1]]))
Model.evaluate(df[df.columns[:-1]],df[df.columns[-1]])
```
